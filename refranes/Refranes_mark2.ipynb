```python
def abre(x):
    with open(x, "r", encoding="utf-8") as archivo:
        refranero= archivo.read()
        refranero= refranero.split("\n")
        archivo.close()
    return refranero
```

```python
import re
def nexus (x):
    lista=[]
    with open(x, "r", encoding="utf-8") as archivo:
        refranero= archivo.read()
        refranero= refranero.split("\n")
        archivo.close()
    with open("nexos.txt", "r", encoding="utf-8") as archivo:
        variables= archivo.read()
        variables= variables.split("\n")
        archivo.close()
    for y in variables:
        expresion_regular=re.compile(r""+" "+y+" ")
        matching= [z for z in refranero if expresion_regular.search(z)]
        nn=y+" "+x
        lista=lista+matching
        if matching== []:
            continue
        else:
            file = open (nn, "w", encoding= "utf-8")
            for q in matching:
                file.write(q+"\n")
    return lista
```
